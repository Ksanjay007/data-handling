```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
    "value":[10,15,22,28,35]
}
df = pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("data")
plt.ylabel("values")
plt.title("time series with thread")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01"]),
    "value":[10,20,17,24,16,18,15,20,17]
    }
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("Date")
plt.ylabel("values")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
text_data="The present is determined by our past actions, and the future by the present."
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('wordcloud')
plt.axis('off')
plt.show()
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('histogram of word frequencies')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show
```
