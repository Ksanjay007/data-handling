```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Subplot 1
x1 = [1, 2, 3]
y1 = [4, 5, 5]
plt.subplot(2, 3, 1)
plt.plot(x1, y1)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 2
x2 = [100, 200, 300, 400]
mylabels = ["Cars", "Sports Bike", "Hyper Naked Bikes", "Scooters"]
mycolors = ["blue", "red", "brown", "pink"]
myexplode = [0.1, 0.1, 0.1, 0.1]
plt.subplot(2, 3, 2)
plt.pie(x2, labels=mylabels, colors=mycolors, explode=myexplode, shadow=9)
plt.title("Pie Chart")

# Subplot 3
x3 = [1, 2, 3, 4, 5]
y3 = [2, 4, 5, 8, 10]
colors3 = ["red", "green", "blue", "yellow", "purple"]
plt.subplot(2, 3, 3)
plt.barh(x3, y3, color=colors3)
plt.title("Horizontal Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 4
plt.subplot(2, 3, 4)
plt.bar(x3, y3, color=colors3)
plt.title("Vertical Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 5
plt.subplot(2, 3, 5)
plt.scatter(x3, y3, s=[1000, 2000, 300, 4000, 500], c=['red', 'green', 'blue', 'yellow', 'purple'], marker="*", edgecolors="black", alpha=1)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the subplots
plt.show()
```
